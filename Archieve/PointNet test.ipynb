```python
# Create a DataFrame for plotting.
df = pd.DataFrame(points, columns=["x", "y", "z"])
# Convert labels to string so Plotly treats them as discrete categories.
df["label"] = labels.astype(str)

# Create a 3D scatter plot with discrete colors.
fig = px.scatter_3d(
    df,
    x="x",
    y="y",
    z="z",
    color="label",  # Discrete color mapping based on the string labels.
    opacity=0.9,
    color_discrete_sequence=px.colors.qualitative.Vivid  # Choose a discrete color set.
)

# Update the marker size to be smaller.
fig.update_traces(marker=dict(size=3))

# Remove axes and backgrounds for a clean visualization.
fig.update_layout(
    scene=dict(
        xaxis=dict(visible=False, showbackground=False, showticklabels=False),
        yaxis=dict(visible=False, showbackground=False, showticklabels=False),
        zaxis=dict(visible=False, showbackground=False, showticklabels=False),
        bgcolor='rgba(0,0,0,0)'  # Transparent 3D scene background.
    ),
    # paper_bgcolor='rgba(0,0,0,0)',  # Transparent outer background.
    # plot_bgcolor='rgba(0,0,0,0)',     # Transparent plot area.
    # Position the legend (the discrete label bar) separately.
    legend=dict(
        x=0.6,   # Moves the legend to the right of the plot.
        y=0.5,
        bgcolor='rgba(0,0,0,0)',
        bordercolor='rgba(0,0,0,0)'
    )
)

# Display the plot.
fig.show()
```
